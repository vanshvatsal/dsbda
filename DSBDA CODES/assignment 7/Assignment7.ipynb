```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LINE PLOT ON AIRQUALITY
airquality = pd.read_csv('airquality.csv')
x = airquality['Temp']
y = airquality['Solar.R']
z = airquality['Wind']

plt.xlabel('Temperature')
plt.ylabel('Solar Radiation')
plt.title('Temperature vs Solar Radiation')
plt.plot(x, y, color='blue', label='2023', marker='o', mfc='red', ls='--', ms=7)
plt.plot(x, z, color='black', label='2024', marker='o', mfc='yellow', ls='--', ms=7)
plt.grid()
plt.legend()
plt.show()
```

```python
#LINE PLOT ON AIRQUALITY
x = airquality['Solar.R']
z = airquality['Temp']

plt.xlabel('Solar Radiation')
plt.ylabel('Temperature')
plt.title('Line Plot')
plt.plot(x, z, color='black', marker='o', mfc='yellow',
         ls='--', ms=7)
plt.grid()
plt.show()
```

```python
#BAR GRAPH ON HEART
heart = pd.read_csv('Heart (1).csv')
chest_pain_types = heart['ChestPain'].unique()
Counts = heart['ChestPain'].value_counts()

# Plotting the bar graph
plt.xlabel('Chest Pain Types')
plt.ylabel('No. of Patients')
plt.title('Distribution of Chest Pain Types', fontsize=20)
plt.bar(chest_pain_types, Counts, color='black', width=0.3, edgecolor='red', linewidth=3, ls=':', alpha=0.5)
plt.grid()
plt.show()
```

```python
#SUBPLOTS ON AIRQUALITY
x = airquality.index
y = airquality['Wind']
z = airquality['Temp']

plt.figure(figsize=(16,10))

plt.subplot(2,2,2)
plt.xlabel('Index')
plt.ylabel('Wind')
plt.title('Wind Speed')
plt.plot(x, y, color='blue', marker='o', mfc='red',
         ls='--', ms=7)

plt.subplot(2,2,1)
plt.xlabel('Index')
plt.ylabel('Terature')
plt.title('Terature Variation')
plt.plot(x, z, color='black', marker='o', mfc='yellow',
         ls='--', ms=7)

plt.tight_layout()
plt.show()
```

```python
#SCATTER PLOT ON HEART
x = heart['Age']
y = heart['RestBP']

plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.scatter(x, y, marker='>')
plt.grid()
plt.show()
```

```python
#PIE CHART ON HEART
column_name_heart = 'ChestPain'  # Change this to the column name you want to plot
students_heart = heart[column_name_heart].value_counts()
courses_heart = students_heart.index  

plt.figure(figsize=(8, 8))
mycolor = ['cyan', 'green', 'blue', 'skyblue', 'red']
plt.title('Pie Chart for {}'.format(column_name_heart))
plt.pie(students_heart, labels=courses_heart, autopct='%.2f%%',
        startangle=90, labeldistance=1.1, radius=0.8,
        colors=mycolor, shadow=True)
plt.show()
```

```python
#BOX PLOT ON HEART
plt.grid()
plt.boxplot(heart['Age'])
plt.title('Boxplot of Age in heart dataset')
plt.ylabel('Age')
plt.show()
```

```python
#CALCULATING INTRQUARTILE RANGE FOR HEART
column_name = 'Age'  # Replace 'Age' with the column name you want to use
grades = heart[column_name]
# Calculate quartiles and IQR
q3 = np.percentile(grades, 75)
q1 = np.percentile(grades, 25)
iqr = q3 - q1
# Calculate upper and lower bounds
upper_bound = q3 + (1.5 * iqr)
lower_bound = q1 - (1.5 * iqr)
# Plot the boxplot
plt.boxplot(grades)
plt.title('Boxplot of ' + column_name)
plt.show()
```

```python
#PLOTS ON AIRQUALITY
x = airquality['Wind']
y = airquality['Temp']
z = airquality['Solar.R']

plt.figure(figsize=(16,10))

plt.subplot(2,2,1)
plt.xlabel('Wind') 
plt.ylabel('Temperature')
plt.title('Line Plot')
plt.plot(x, y, color='blue', label='2023', marker='o', mfc='red',
         ls='--', ms=7)
plt.plot(x, z, color='black', label='2024', marker='o', mfc='yellow',
         ls='--', ms=7)
plt.grid()
plt.legend()

plt.subplot(2,2,2)
courses = ['Ozone', 'Solar.R', 'Wind', 'Temp']
students = [len(airquality['Ozone']), len(airquality['Solar.R']),
            len(airquality['Wind']), len(airquality['Temp'])]
plt.xlabel('Variables')
plt.ylabel('Count')
plt.title('Bar Graph', fontsize=20)
plt.bar(courses, students, color='black', width=0.3,
        edgecolor='red', linewidth=3, ls=':', alpha=0.5)
plt.grid()

plt.subplot(2,2,3)
plt.xlabel('Wind')
plt.ylabel('Terature')
plt.scatter(x, y, marker='>')
plt.grid()

plt.subplot(2,2,4)
mycolor = ['cyan', 'green', 'blue', 'skyblue']
plt.title('Pie Chart')
plt.pie(students, labels=courses, autopct='%.2f%%',
        startangle=90, labeldistance=1.1, radius=0.8,
        colors=mycolor, shadow=True)

plt.savefig('MY FIRST PLOT')
plt.show()
```
